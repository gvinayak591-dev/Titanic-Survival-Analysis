# Titanic Data Analysis 🚢

## Objective
To analyze survival patterns of passengers based on class and age using Python, Pandas, and Matplotlib.

## Tools Used
- Python
- Pandas
- Matplotlib

## Dataset
## Dataset
The Titanic dataset is a real-world dataset commonly used in machine learning for classification problems.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Handle missing values
df["Age"] = df["Age"].fillna(df["Age"].mean())
# -------------------------------
# Survival by Class
# -------------------------------
df.groupby("Pclass")["Survived"].mean().plot(kind="bar")
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()

# -------------------------------
# Survival by Age Group
# -------------------------------
df["AgeGroup"] = pd.cut(
    df["Age"],
    bins=[0, 12, 18, 40, 60, 100],
    labels=["Child", "Teen", "Adult", "Mid Age", "Old"]
)

df.groupby("AgeGroup", observed=True)["Survived"].mean().plot(kind="bar")
plt.title("Survival Rate by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Survival Rate")
plt.show()

# -------------------------------
# Survival by Gender
# -------------------------------
df.groupby("Sex")["Survived"].mean().plot(kind="bar")
plt.title("Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.show()
```

## Insights 📊

1. Passengers in **1st class had the highest survival rate**, showing that socio-economic status played a major role.

2. **3rd class passengers had the lowest survival rate**, indicating limited access to safety resources.

3. **Children had better survival chances** compared to adults, possibly due to priority rescue.

4. Survival rate decreases as we move from **higher class to lower class**.

5. Middle-aged and older passengers had relatively **lower survival rates** compared to younger groups.

6. Both **age and class significantly influenced survival probability**.
## Conclusion 🎯

This analysis shows that survival on the Titanic was not random.  
Factors like **passenger class and age** had a strong impact on survival chances.
